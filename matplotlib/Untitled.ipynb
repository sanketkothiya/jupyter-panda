# matplotlib tutorial

```python
import numpy as np
from matplotlib import pylab as plt
```

```python
x=np.linspace(0,10,25)
y=x*x + 2
# print(x)
# print(y)
print(np.array([x,y]).reshape(25,2).reshape(2,25))
```

```python
plt.plot(x,y,'r')
#  we use r,c,k,b
```

```python
plt.subplot(1,2,1) # this is (row,columns,index)
plt.plot(x,y,'r--') #h ere r-- is line style
plt.subplot(1,2,2)
plt.plot(x,y,'g*--')
```
