# IPL 2020 DATA


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# READ CSV FILE

```python
df=pd.read_csv("ipl2020.csv")
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

# CHANGE COLUMN NAME

```python
df.columns=['index', 'ListNO', 'SetNo.', '2020Set', 'FirstName',
       'Surname', 'Country', 'State_Association', 'DOB', 'Age', 'Specialism',
       'Unnamed: 10', 'position', 'Testcaps', 'ODIcaps', 'T20caps',
       'IPL', 'Previous IPLTeam(s)', '2019Team', '2019IPL', 'C/U/A',
       'ReservePrice', 'Surname.1']
```

```python
df.head(2)
```

# Drop columns


```python
df.drop(['Testcaps','Unnamed: 10',"Surname.1","2019Team","2019IPL","C/U/A","ListNO"],axis=1,inplace=True)
```

```python
df
```

# REPLACE COLUMNS  HEADER BY UPPERCASE

```python
df.columns=[x.upper() for x in df.columns]
```

```python
df.head()
```

# REPLACE NAN VALUE BY VALID VALUES

```python
new_df=df.isna()
new_df
```

```python
new_df=df.fillna({
    'IPL':0,
    "T20CAPS":0,
    "ODICAPS":0,
    "PREVIOUS IPLTEAM(S)":"no-data"
    
    
})
```

```python
new_df.to_csv('removeipl2020',index=False)
```

```python
new_df
```

# FILTERING THE DATA

```python
filter_data=new_df.loc[ (df['SPECIALISM'] == 'BATSMAN')  & (df['IPL'] >= 70)]
```

```python
filter_data
```

```python
filter1_data=new_df.loc[ (df['T20CAPS'] >= 70)]
```

```python
filter1_data
```

# GRAPH

# 1. batsman highest score in ipl

```python
plt.plot(filter_data.FIRSTNAME,filter_data.IPL, color='red', marker='o')
plt.title('players performance in ipl', fontsize=14)
plt.xlabel('player', fontsize=14)
plt.ylabel('score', fontsize=14)
plt.grid(True)
plt.show()
```

#  2. batsman highest score in T20 

```python
plt.style.use('ggplot')
plt.bar(filter1_data.FIRSTNAME.head(6),filter1_data.T20CAPS.head(6), color='green')
plt.xlabel("players name")
plt.ylabel("T20 score")
plt.title("player performance in T20")
plt.show()
```
