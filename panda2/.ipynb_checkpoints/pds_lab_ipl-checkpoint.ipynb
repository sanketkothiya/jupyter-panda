# REPLACE THE NAN VALUE

```python
new_df=df.isna()
new_df
```

```python
nan_change=df.fillna(0)
nan_change.head(100)
```

```python
new_df=df.fillna({
    'IPL':0,
    "PREVIOUS IPLTEAM(S)":0,
    "T20\nCAPS":0,
    "ODI\nCAPS":0
    
    
})
```

```python
new_df.to_csv('nanremove',index=False)
```

```python
new_df.head(10)
```

```python

  
plt.plot(new_df.FIRSTNAME.head(7),new_df.IPL.head(7), color='red', marker='o')
plt.title('players performance in ipl', fontsize=14)
plt.xlabel('player', fontsize=14)
plt.ylabel('score', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.style.use('ggplot')
plt.bar(new_df.FIRSTNAME.head(6),new_df.T20CAPS.head(6), color='green')
plt.xlabel("player name")
plt.ylabel("T20 score")
plt.title("player performance in T20")
plt.show()
```
